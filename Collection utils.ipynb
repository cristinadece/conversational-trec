# Create collection file

```python
# iterate over files in that directory
# save new info to file
# delete file after processing to save space

doc_count=0 
directory = "/data3/muntean/DRhard/data/passage/dataset/CAST-2022/jsonlines"
with open("/data3/muntean/DRhard/data/passage/dataset/CAST-2022/CAST2022collection.passage.tsv", 'w') as output:
    dir_list = os.listdir(directory)
    for filename in dir_list:
        f = os.path.join(directory, filename)
        # checking if it is a file
        if os.path.isfile(f):
            print(f)
            with open(f, 'r') as file:
                for line in file:
                    doc_count+=1
                    # print(line)
                    a = json.loads(line)
                    for passage in a['contents']:
                        contents = passage['body'].replace('\n', ' ').replace('\t', ' ')
                        output.write(a['id']+"-"+str(passage['id']) + '\t' + contents + '\n')
        os.remove(f)
```
